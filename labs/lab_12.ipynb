Pull [Wikipedia's list of countries by area](https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_area#Countries_and_dependencies_by_area) into a DataFrame using `read_html()`.


```python
import pandas as pd
import requests
import urllib3

urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

url = "https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_area#Countries_and_dependencies_by_area"

headers = {"User-Agent": "Mozilla/5.0"}
response = requests.get(url, headers=headers, verify=False)
html = response.text

df_list = pd.read_html(html)

df = df_list[1]
df.head()
```

## Time series

Using [MTA ridership data](https://data.ny.gov/Transportation/MTA-Daily-Ridership-Data-Beginning-2020/vxuj-8kew/about_data), **view the bus ridership over time**. Start with per day, then do per week.


```python
df = pd.read_csv("MTA_Daily_Ridership_Data__2020_-_2025_20251121.csv")
df.head()
```

```python
df = df[["Date", "Buses: Total Estimated Ridership"]]
df.head()
```

```python
df.info()
```

```python
df["Date"] = pd.to_datetime(df["Date"], format="%m/%d/%Y")
df.head()
```

```python
df.info()
```

```python
import plotly.express as px
fig = px.line(df, x="Date", y="Buses: Total Estimated Ridership", title="MTA Daily Bus Ridership (2020-2025)")
fig
```

```python
weekly = df.resample("W", on="Date").sum().reset_index()
weekly
```

```python
fig = px.line(
    weekly,
    x="Date",
    y="Buses: Total Estimated Ridership",
    title="MTA Weekly Bus Ridership (2020–2025)"
)
fig
```

## API

You're going to pick a [World Bank Indicator](https://data.worldbank.org/indicator) and plot that over time. You'll retrieve the data through [the API](https://datahelpdesk.worldbank.org/knowledgebase/articles/889392-about-the-indicators-api-documentation).


### Tips

- Start simple
- Use the Python [requests](https://requests.readthedocs.io/) package
- See [`json_normalize()`](https://pandas.pydata.org/docs/reference/api/pandas.json_normalize.html)

Retrieve _all_ data for that Indicator through [the API](https://datahelpdesk.worldbank.org/knowledgebase/articles/889392-about-the-indicators-api-documentation) — see "Page".


```python
import requests
url = " https://api.worldbank.org/v2/country/all/indicator/SP.POP.TOTL?format=json"
response = requests.get(url)
response
```

```python
response.json()
```

```python
df = pd.json_normalize(response.json()[1])
df.head()
```

### Pagination

You'll probably want to create DataFrames for each page, then "concatenate" them. Here's a structure you can start with:


```python
# in a loop
#     get the first/next page of data
#     combine with the data that's already been retrieved
#     if there are fewer than the default number of records returned, stop the loop
```

```python
total_pages = response.json()[0]['pages']
total_pages
```

```python
for i in range(2, total_pages + 1):
    print("Processing page:", i)
    url = f"https://api.worldbank.org/v2/country/all/indicator/SP.POP.TOTL?format=json&page=" + str(i)
    response = requests.get(url)
    df_next = pd.json_normalize(response.json()[1])
    df = pd.concat([df, df_next], ignore_index=True)

df.info()
```
